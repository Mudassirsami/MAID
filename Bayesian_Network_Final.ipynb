```python
from pgmpy.models import BayesianModel
from pgmpy.factors.discrete import TabularCPD
import networkx as nx
import matplotlib.pyplot as plt

# Define the structure of the Bayesian network
model = BayesianModel([('H', 'E1'), ('H', 'E2'), ('H', 'E3'), ('H', 'E4')])

# Define the conditional probability distributions for each node
cpd_H = TabularCPD(variable='H', variable_card=2, values=[[0.063], [0.937]])
print("Conditional Probability Distribution for Node H:")
print(cpd_H)

cpd_E1 = TabularCPD(variable='E1', variable_card=2,
                    values=[[0.9492, 0.0507],
                            [0.1609, 0.8340]],
                    evidence=['H'], evidence_card=[2])
print("\nConditional Probability Distribution for Node E1:")
print(cpd_E1)

cpd_E2 = TabularCPD(variable='E2', variable_card=2,
                    values=[[0.8857, 0.1142],
                            [0.5921, 0.4078]],
                    evidence=['H'], evidence_card=[2])
print("\nConditional Probability Distribution for Node E2:")
print(cpd_E2)

cpd_E3 = TabularCPD(variable='E3', variable_card=2,
                    values=[[0.6412, 0.3587],
                            [0.3271, 0.6729]],
                    evidence=['H'], evidence_card=[2])
print("\nConditional Probability Distribution for Node E3:")
print(cpd_E3)

cpd_E4 = TabularCPD(variable='E4', variable_card=2,
                    values=[[0.62, 0.37],
                            [0.4292, 0.5707]],
                    evidence=['H'], evidence_card=[2])
print("\nConditional Probability Distribution for Node E4:")
print(cpd_E4)

# Add the conditional probability distributions to the model
model.add_cpds(cpd_H, cpd_E1, cpd_E2, cpd_E3, cpd_E4)

# Create a directed graph
graph = nx.DiGraph()

# Add nodes to the graph for H1 and H2 in the first row
graph.add_node('H1', pos=(1, 1))
graph.add_node('H2', pos=(3, 1))

# Add nodes to the graph for E1, E2, E3, and E4 in the second row
for i, node in enumerate(['E1', 'E2', 'E3', 'E4']):
    graph.add_node(node, pos=(i+1, 0))
    # Add the conditional probability table as node attribute
    graph.nodes[node]['cpd'] = model.get_cpds(node).to_factor().values

# Add edges to the graph
for node in ['E1', 'E2', 'E3', 'E4']:
    graph.add_edge('H1', node)
    graph.add_edge('H2', node)

# Draw the graph with conditional probability tables as labels
pos = nx.get_node_attributes(graph, 'pos')
nx.draw(graph, pos, with_labels=False, node_size=3000, node_color='skyblue', font_size=12, font_weight='bold')

# Add labels for conditional probability tables
for node, (x, y) in pos.items():
    if 'cpd' in graph.nodes[node]:
        table = graph.nodes[node]['cpd']
        table_str = "\n".join([" ".join(map(str, row)) for row in table])
        plt.text(x, y -0.2, f"CPD:\n{table_str}", horizontalalignment='left', verticalalignment='center')

# Add labels for nodes
nx.draw_networkx_labels(graph, pos, labels={node: node for node in graph.nodes()}, font_weight='bold')

plt.title("Bayesian Network")
plt.show()
```

# **Perform prediction inference queries**

```python
from pgmpy.models import BayesianNetwork
from pgmpy.factors.discrete import TabularCPD
import networkx as nx
import matplotlib.pyplot as plt

# Define the structure of the Bayesian network
model = BayesianNetwork([('H', 'E1'), ('H', 'E2'), ('H', 'E3'), ('H', 'E4')])

# Define the conditional probability distributions for each node
cpd_H = TabularCPD(variable='H', variable_card=2, values=[[0.063], [0.937]])
```

```python
# Fix CPD for E1
cpd_E1 = TabularCPD(variable='E1', variable_card=2,
                    values=[[0.9492, 0.0508],
                            [0.1609, 0.8391]],
                    evidence=['H'], evidence_card=[2])

# Normalize the values for E1
cpd_E1.normalize()

# Fix CPD for E2
cpd_E2 = TabularCPD(variable='E2', variable_card=2,
                    values=[[0.8857, 0.1143],
                            [0.5922, 0.4078]],
                    evidence=['H'], evidence_card=[2])

# Normalize the values for E2
cpd_E2.normalize()

# Fix CPD for E3
cpd_E3 = TabularCPD(variable='E3', variable_card=2,
                    values=[[0.6412, 0.3588],
                            [0.3272, 0.6728]],
                    evidence=['H'], evidence_card=[2])

# Normalize the values for E3
cpd_E3.normalize()

# Fix CPD for E4
cpd_E4 = TabularCPD(variable='E4', variable_card=2,
                    values=[[0.62, 0.38],
                            [0.4292, 0.5708]],
                    evidence=['H'], evidence_card=[2])

# Normalize the values for E4
cpd_E4.normalize()
```

```python
# Add the conditional probability distributions to the model
model.add_cpds(cpd_H, cpd_E1, cpd_E2, cpd_E3, cpd_E4)
```

```python
from pgmpy.inference import VariableElimination
```

```python
# Create a VariableElimination object with the Bayesian network model
infer = VariableElimination(model)
```

```python
# Perform prediction inference query for P(E1 | H=1)
query_result = infer.query(variables=['E1'], evidence={'H': 1})
print("P(E1 | H=1):")
print(query_result)

# Perform prediction inference query for P(E2 | H=0)
query_result = infer.query(variables=['E2'], evidence={'H': 0})
print("P(E2 | H=0):")
print(query_result)

# Perform prediction inference query for P(H | E3=1, E4=0)
query_result = infer.query(variables=['H'], evidence={'E3': 1, 'E4': 0})
print("P(H | E3=1, E4=0):")
print(query_result)
```
