# DataFrame

```python
# Create a DataFrame from the provided data
classifier_data = {
    'Model': ['Linear', 'KNN', 'SVM', 'XGB', 'Decision Tree', 'Random Forest'],
    'MSE': [106.25, 140.27, 159.02, 0.0, 0.0, 0.0],
    'RMSE': [1.03, 1.18, 1.26, 0.0, 0.0, 0.0],
    'Accuracy': [0.67, 0.62, 0.54, 1.0, 1.0, 1.0],
    'F1_Score': [0.31, 0.29, 0.20, 1.0, 1.0, 1.0]
}
```

```python
df_classifier = pd.DataFrame(classifier_data)
```

```python
# Set the Model column as the index for better visualization
df_classifier.set_index('Model', inplace=True)

# Create a color map for different metrics
colors = ['blue', 'green', 'orange', 'red']

# Plot the data
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
fig.suptitle('Performance Metrics by Model', fontsize=16)

# MSE plot
df_classifier[['MSE']].plot(kind='bar', ax=axes[0, 0], color=colors[0])
axes[0, 0].set_title('Mean Squared Error (MSE)')

# RMSE plot
df_classifier[['RMSE']].plot(kind='bar', ax=axes[0, 1], color=colors[1])
axes[0, 1].set_title('Root Mean Squared Error (RMSE)')

# Accuracy plot
df_classifier[['Accuracy']].plot(kind='bar', ax=axes[1, 0], color=colors[2])
axes[1, 0].set_title('Accuracy')

# F1 Score plot
df_classifier[['F1_Score']].plot(kind='bar', ax=axes[1, 1], color=colors[3])
axes[1, 1].set_title('F1 Score')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
